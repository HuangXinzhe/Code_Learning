```python
from sklearn.datasets import load_breast_cancer#乳腺癌数据集
```

```python
cancer= load_breast_cancer()
```

```python
X = cancer.data
```

```python
y = cancer.target
```

```python
X.shape
```

```python
X[0]
```

```python
from sklearn.preprocessing import StandardScaler 
```

```python
help(StandardScaler)
```

```python
x_std = StandardScaler()
```

```python
X_std = x_std.fit(X)
```

```python
X=X_std.transform(X)
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y,test_size=0.2)#切分数据集
```

#### 1.逻辑回归

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)#训练模型
test_score = model.score(X_test, y_test)#计算r2得分
print('test_score:',test_score)
```

```python
help(model.score)
```

#### 2.svm支持向量机

```python
from sklearn.svm import SVC
```

```python
model =SVC(kernel="linear", C=0.025)
#kernel:核函数种类，C惩罚因子。C: 惩罚系数，用来控制损失函数的惩罚系数，类似于LR中的正则化系数。
#C越大，相当于惩罚松弛变量，希望松弛变量接近0，即对误分类的惩罚增大，趋向于对训练集全分对的情况，
#这样会出现训练集测试时准确率很高，但泛化能力弱，容易导致过拟合。 C值小，对误分类的惩罚减小，容错能力增强，
#泛化能力较强，但也可能欠拟合。
model.fit(X_train, y_train)
test_score = model.score(X_test, y_test)
print('test_score:',test_score)
```

```python
# model =SVC(gamma=2,C=1)
#gamma：核函数的系数，C惩罚因子。核函数系数，该参数是rbf，poly和sigmoid的内核系数；默认是'auto'，
#那么将会使用特征位数的倒数，即1 / n_features。（即核函数的带宽，超圆的半径）。gamma越大，σ越小，
#使得高斯分布又高又瘦，造成模型只能作用于支持向量附近，可能导致过拟合；反之，gamma越小，σ越大，
#高斯分布会过于平滑，在训练集上分类效果不佳，可能导致欠拟合。 
model =SVC()
model.fit(X_train, y_train)
test_score = model.score(X_test, y_test)
print('test_score:',test_score)
```

#### 3.决策树

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# model =DecisionTreeClassifier(max_depth=5)#max_depth决策树的最大深度。
model =DecisionTreeClassifier(max_depth=5)
model.fit(X_train, y_train)
test_score = model.score(X_test, y_test)
print('test_score:',test_score)
```

#### 4.随机森林

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(max_depth=5, n_estimators=10)
##max_depth决策树的最大深度。n_estimators：决策树的个数。
model.fit(X_train, y_train)
test_score = model.score(X_test, y_test)
print('test_score:',test_score)
```

#### 5.Adaboost

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
model = AdaBoostClassifier()
model.fit(X_train, y_train)
test_score = model.score(X_test, y_test)
print('test_score:',test_score)
```

#### 6.神经网络

```python
from sklearn.neural_network import MLPClassifier
```

```python
model = MLPClassifier(alpha=1, max_iter=1000)
#alpha:L2正则化系数，max_iter:最大迭代次数
model.fit(X_train, y_train)
test_score = model.score(X_test, y_test)
print('test_score:',test_score)
```
