Python is a multi-paradigm programming language. It supports different programming approaches.

One of the popular approaches to solve a programming problem is by creating objects. This is known as Object-Oriented Programming (OOP).

```python
class Person: # class: 类
    """
    attributes/properties: name, age, job, marrige, school
    methods: 
    walk, sleep, sell, print_name,...
    """
    def __init__(self, name, age): # class constructor, 构造函数
        self.name = name
        self.age = age

    def print_name(self):
        print("Hello my name is " + self.name)   
    

p1 = Person("John", 36)
p2 = Person("Jack", 30)
p3 = Person("Mike", 25)

p1.print_name()
p2.print_name()
p3.print_name()
```

```python
# 班级管理系统
# 一个院系有多个班级
# 每个班级有学生，
# 每个学生有自己的信息，学生的成绩
# 每个学生的选课

# 一个院系有多个班级，一个班级多个学生， 每个学生只上一门课

# 设计角度：院系的class, 班级的class, 学生的class


class Institute:
    
    def __init__(self, institute_name):
        self.institute_name = institute_name
        
        self.groups = []
    
    def add_group(self, group):
        self.groups.append(group)
     
    def delele_group(self, group_name):
        pass
    
    def get_count(self):
        stu_cnt = 0
        for group in self.groups:
            stu_cnt = stu_cnt + group.get_student_count()
        
        return stu_cnt
    
    def get_average_score(self):
        # 80.  40
        # 78.  30
        # (80*40+78*30) / (40+30)
        avg_score = 0
        for group in self.groups:
            avg_score = group.get_student_count() * group.avg_record()
        
        return avg_score/self.get_count()
        
    def get_pass_rate(self):
        pass
    
    
class Group:
    def __init__(self, group_name):
        self.group_name = group_name
        
        self.students = []
    
    def avg_record(self):
        """
        this function return the average score of the students 
        in this group
        
        return average score: float
        """
        record_sum = 0
        for i in range(len(self.students)):
            record_sum = record_sum + self.students[i].get_record()
        
        return record_sum/len(self.students)
        
    
    def get_student_count(self):
        """
        return the number of students in this group
        """
        
        return len(self.students)
    
    
    def pass_rate(self, pass_score):
        count = len(self.students)
        pass_cnt = 0
        for i in range(count):
            if self.students[i].get_record() >= pass_score:
                pass_cnt = pass_cnt+1
        
        return pass_cnt/count
        
    
    def add_to_group(self, stu):
        self.students.append(stu)
    
    def add_record(self, stu_id, score):
        stu_idx = -1
        for i in range(len(self.students)):
            if self.students[i].get_id() == stu_id:
                stu_idx = i
                break
        if stu_idx >=0:
            self.students[stu_idx].set_record(score)
        
    
    def delete_from_group(self, user_id):
        # search for student whose id is "user_id"
        stu_idx = -1
        for i in range(len(self.students)):
            if self.students[i].get_id() == user_id:
                stu_idx = i
                break
        
        if stu_idx >= 0: # cannot search for the id 
            self.students.delete(stu_idx) # delete specific item from the list indexed by stu_idx
        
        
    
class Student:
    
    def __init__(self, user_id, name, age, address):
        self.user_id = user_id
        self.name = name
        self.age = age
        self.address = address
        
        self.record = -1 # float
    
    def print_student_info():
        print("student info:",user_id, name, age, address)
    
    def set_record(self, record):
        self.record = record
        
    def get_record(self):
        return self.record
    
    def update_address(self, address):
        self.address = address
        
    def get_id(self):
        return self.user_id
    
    def print_record(self):
        if (self.record < 0):
            print ("用户的成绩暂时不存在！")
        else:
            print ("用户的成绩为：%f", self.record)
    
```
