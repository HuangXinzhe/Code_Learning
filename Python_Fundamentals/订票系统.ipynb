```python
## 定义用户已经订的车票变量
reserve_ticket_info = {}

while True:
    # 用户进入系统后，提供欢迎语，并告知每个编号的作用
    business_no = input("欢迎使用订票系统，请输入您想要办理的业务。0:列车时刻查询;1:已订票查询；2:订票；3：退票；4：改签；9：退出系统。请选择：")
    
    # 当用户输入0时，展示出我们的列车时刻表，注意要打印成我们图中展示的样子
    if business_no == "0":
        # 先打印一下表头，其中%-10s的意思是格式化字符串的长度为10
        # 当原字符串长度小于10时，用空格补齐
        # 这样做的目的是让后边的打印也对齐
        print("%-10s%-10s%-10s%-10s%-10s%-10s" %("车次","始发站","到达站","出发时间","到达时间","历时"))
        # 遍历列表，取出其中每一个元素，就是里边的字典，然后再遍历字典的值
        
        # len函数的作用是获取到这个列表的长度
        # range函数的意思是从0开始生成数字，一直到我们给定长度的这个
        # 例如：range(3),那么生成的数字列表就是0,1,2
        for i in range(len(train_time_list)):
            print("%-10s%-10s%-10s%-10s%-10s%-10s" %(train_time_list[i].get("车次"),
                                                     train_time_list[i].get("始发站"),
                                                     train_time_list[i].get("到达站"),
                                                     train_time_list[i].get("出发时间"),
                                                     train_time_list[i].get("到达时间"),
                                                     train_time_list[i].get("历时")))

    # 当用户选择1的时候，就是已订票查询，需要先检查用户是否已经订票了
    elif business_no == "1":
        if reserve_ticket_info != {}:
            print("{}您好,您已经预定{}次的列车,{}开，始发站是{},到达站是{},到达时间是{},历时{},请您尽快换取纸质车票。".format(
                reserve_ticket_info.get("姓名"),
                reserve_ticket_info.get("车次"),
                reserve_ticket_info.get("始发站"),
                reserve_ticket_info.get("到达站"),
                reserve_ticket_info.get("出发时间"),
                reserve_ticket_info.get("到达时间"),
                reserve_ticket_info.get("历时")
            ))
        else:
            print("您还没有预定车票哦")
    # 2是让用户选择订票，让用户输入姓名和车次
    elif business_no == "2":
        name = input("请输入购票车姓名:")
        # 输入购买车次的时候需要检查车次是否存在
        while True:
            # 获取用户输入的购买车次
            user_train_no = input("请输入购买车次:")
            # 检查用户输入车次是否存在在我们的列表时刻列表里，如果不存在就请重新输入
            system_train_no_list = []
            for i in range(len(train_time_list)):
                system_train_no_list.append(train_time_list[i].get("车次"))
        
            if user_train_no.upper() not in system_train_no_list:
                print("对不起，您输入的车次不存在")
                continue
            else:
                reserve_ticket_info["姓名"] = name
                # 获取到车次的详细信息并记录到reserve_ticket_info中
                for i in range(len(train_time_list)):
                    if train_time_list[i].get("车次") == user_train_no.upper():
                        reserve_ticket_info['车次'] = train_time_list[i].get("车次")
                        reserve_ticket_info['始发站'] = train_time_list[i].get("始发站")
                        reserve_ticket_info['到达站'] = train_time_list[i].get("到达站")
                        reserve_ticket_info['出发时间'] = train_time_list[i].get("出发时间")
                        reserve_ticket_info['到达时间'] = train_time_list[i].get("到达时间")
                        reserve_ticket_info['历时'] = train_time_list[i].get("历时")
                print("{}您好,您已经预定{}次的列车,始发站是{},到达站是{},出发时间是{},到达时间是{},历时{},请您尽快换取纸质车票。".format(
                    reserve_ticket_info.get("姓名"),
                    reserve_ticket_info.get("车次"),
                    reserve_ticket_info.get("始发站"),
                    reserve_ticket_info.get("到达站"),
                    reserve_ticket_info.get("出发时间"),
                    reserve_ticket_info.get("到达时间"),
                    reserve_ticket_info.get("历时")
                    ))
                break
    # 退订逻辑：首先检查用户是否订票，未订票就让用户先去订票
    # 如果订票了，先展示订票信息，然后确认是否退订
    elif business_no == "3":
        if reserve_ticket_info == {}:
            print("您还未预购车票，请先预购车票")
        else:
            cancel_ticket_status = input("{}您好,您已经预定{}次的列车,始发站是{},到达站是{},出发时间是{},到达时间是{},历时{},确认退订请按Y，后悔了就按N，请按吧:".format(
                    reserve_ticket_info.get("姓名"),
                    reserve_ticket_info.get("车次"),
                    reserve_ticket_info.get("始发站"),
                    reserve_ticket_info.get("到达站"),
                    reserve_ticket_info.get("出发时间"),
                    reserve_ticket_info.get("到达时间"),
                    reserve_ticket_info.get("历时")
                    ))
            if cancel_ticket_status.upper() == "Y":
                reserve_ticket_info = {}
                print("退票成功")
    # 改签逻辑：首先要展示一下所有的车次，改签不能改签自己已经定的车次
    elif business_no == "4":
        if reserve_ticket_info == {}:
            print("您还未预购车票，请先预购车票")
        else:
            for i in range(len(train_time_list)):
                print("%-10s%-10s%-10s%-10s%-10s%-10s" %(train_time_list[i].get("车次"),
                                                     train_time_list[i].get("始发站"),
                                                     train_time_list[i].get("到达站"),
                                                     train_time_list[i].get("出发时间"),
                                                     train_time_list[i].get("到达时间"),
                                                     train_time_list[i].get("历时")))
            while True:
                alter_train_no = input("请输入您要改签的车次:")
                # 判断改签车次是否与已预订车次是同一车次
                if alter_train_no.upper() == reserve_ticket_info.get("车次"):
                    print("改签车次与已预订车次不能是同一车次哦！！！")
                    continue
                # 这里边缺少了，如果车次不存在的判断
                else:
                    for i in range(len(train_time_list)):
                        if train_time_list[i].get("车次") == alter_train_no.upper():
                            reserve_ticket_info['车次'] = train_time_list[i].get("车次")
                            reserve_ticket_info['始发站'] = train_time_list[i].get("始发站")
                            reserve_ticket_info['到达站'] = train_time_list[i].get("到达站")
                            reserve_ticket_info['出发时间'] = train_time_list[i].get("出发时间")
                            reserve_ticket_info['到达时间'] = train_time_list[i].get("到达时间")
                            reserve_ticket_info['历时'] = train_time_list[i].get("历时")
                    print("{}您好,改签成功。您已经预定{}次的列车,始发站是{},到达站是{},出发时间是{},到达时间是{},历时{},请您尽快换取纸质车票。".format(
                        reserve_ticket_info.get("姓名"),
                        reserve_ticket_info.get("车次"),
                        reserve_ticket_info.get("始发站"),
                        reserve_ticket_info.get("到达站"),
                        reserve_ticket_info.get("出发时间"),
                        reserve_ticket_info.get("到达时间"),
                        reserve_ticket_info.get("历时")
                        ))
                    break
    elif business_no == "9":
        print("票务系统很高兴为您服务，欢迎下次光临")
        break
    else:
        print("号码输入有误")
```
